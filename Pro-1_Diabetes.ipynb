```python
import numpy as np
import pandas as pd
```

```python
import tensorflow as tf
from tensorflow import keras
```

### Loading Data set creating samples and labels for training

```python
Data_set = np.loadtxt("Diabetes.csv", delimiter=',')
train_samples = Data_set[:, 0:8]
train_labels = Data_set[:, 8]
```

### Scaling down samples

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_samples = scaler.fit_transform(train_samples)
```

### Creating Model Architecture

```python
model = keras.models.Sequential()
model.add(keras.layers.Dense(8, input_dim=8))
model.add(keras.layers.Dense(64, activation = "relu"))
model.add(keras.layers.Dense(64, activation = "relu"))
model.add(keras.layers.Dense(64, activation = "relu"))
model.add(keras.layers.Dense(1, activation = "sigmoid"))
```

### Compiling the model

```python
model.compile(loss = "binary_crossentropy", optimizer = 'adam', metrics = ['accuracy'])
```

### Defining checkpoints to stop the training at minimum loss and save that model

```python
checkpoint_cb = keras.callbacks.ModelCheckpoint("Diabetes_model", save_best_only=True)
early_stopping_cb = keras.callbacks.EarlyStopping(patience=20, restore_best_weights=True)
```

### Initiating with fixed random seeds to get the same output everytime

```python
np.random.seed(50)
tf.random.set_seed(50)
```

### Training the Model

```python
model.fit(train_samples, train_labels, epochs = 1000, validation_split = 0.1,
          callbacks = [checkpoint_cb, early_stopping_cb])
```

### Creating test data and scaling it

```python
model = keras.models.load_model("Diabetes_model")
test_data = np.loadtxt("data.csv", delimiter = ',')
test_samples = test_data[:, 0:8]
test_labels = test_data[:, 8]
test_samples = scaler.fit_transform(test_samples)
```

### Evaluating our model

```python
model.evaluate(test_samples, test_labels)
```

### Prediction class to make our output meaningful

```python
predict = ["Patient has no risk of developing diabetes", "Patient has risk of developing diabetes within next 5 years"]
```

### Creating a patient record and scaling it

```python
patient_record = test_data[45, 0:8][np.newaxis]
patient_record = scaler.fit_transform(patient_record)
```

### Example of model predicting in a real scenario

```python
predict[int(model.predict(patient_record).round(1))]
```
